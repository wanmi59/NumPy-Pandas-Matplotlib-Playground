```python
data = {'a': np.arange(50),
       'c': np.random.randint(0, 50, 50),
       'd': np.random.randn(50)}

data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c = 'c', s = 'd', data = data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

names = ['group a', 'group b', 'group c']
values = [1, 10, 100]

plt.figure(figsize=(12,4))
plt.gca()
plt.subplot(131)
plt.bar(names, values)

plt.subplot(132)
plt.scatter(names, values)

plt.subplot(133)
plt.plot(names, values, '-*', linewidth = 3, color = 'black', 
         markersize = 20, markerfacecolor = 'blue', markeredgecolor = 'blue')

plt.suptitle('Categorical Plotting')
plt.show()
```

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)
n = plt.hist(x, 50, density = 1, facecolor = 'g', alpha = 0.75)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram')
plt.text(60, .025, r'$\mu-100, \ \sigma =15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)

plt.show()
```

## Countplot

```python
df = pd.DataFrame(np.random.randn(5, 3), columns = list('ABC'))
df['Class'] = np.random.randint(0, 2, (5, 1))

plt.figure(figsize=(8,4))
sns.countplot('Class', data=df, hue='Class')
plt.ylabel('Frequency')
plt.show()
```

## Distribution Plot

```python
df = pd.DataFrame(np.random.randn(10, 3), columns = list('ABC'))
df['Class'] = np.random.randint(0, 2, (10, 1))

plt.figure(figsize=(10,5))
sns.distplot(df['A'][df['Class'] == 0], label='Class 0')
sns.distplot(df['A'][df['Class'] == 1], label='Class 1')
plt.legend()
plt.ylabel('Frequency')
plt.show()
```
